```python
import requests
import pandas as pd
```

```python
async def download(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(filename, "wb") as f:
            f.write(response.content)
```

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv"
await download(file_path, "m1_survey_data.csv")
```

```python
file_name = "m1_survey_data.csv"
df = pd.read_csv(file_name)
```

```python
duplicate_rows = df.duplicated().sum()
duplicate_rows
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
#finding missing values
df.isna().sum()
```

```python
df.isna().sum()['WorkLoc']
```

```python
value_counts_workloc = df['WorkLoc'].value_counts()
value_counts_workloc
```

```python
max_value = value_counts_workloc.idxmax()
df['WorkLoc'].fillna(max_value, inplace=True)
```

```python
df.isna().sum()['WorkLoc']
```

```python
df['CompFreq'].unique()
```

```python
#normalizing data
def normalize_comp(row):
  if row['CompFreq'] == 'Yearly':
    return row['CompTotal']
  elif row['CompFreq'] == 'Monthly':
    return row['CompTotal'] * 12
  elif row['CompFreq'] == 'Weekly':
    return row['CompTotal'] * 52
  else:
    return None

df['NormalizedAnnualCompensation'] = df.apply(normalize_comp, axis=1)
df[['CompFreq', 'CompTotal', 'NormalizedAnnualCompensation']].head()
```
